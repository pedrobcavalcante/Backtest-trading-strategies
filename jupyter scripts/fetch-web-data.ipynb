```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import json
import yfinance as yf
from datetime import date
```

```python
url = "https://www.infomoney.com.br/cotacoes/empresas-b3/"

options = webdriver.ChromeOptions()
options.headless = True
driver = webdriver.Chrome("chromedriver", options=options)

driver.get(url)

rows = driver.find_elements(By.CSS_SELECTOR, ".article-content table tbody tr, table.default-table tbody tr")
list = []
for row in range(0,len(rows)):
    # Get the columns (all the column 2)
    col = driver.find_elements(By.CSS_SELECTOR, ".article-content table tbody td.strong, table.default-table tbody td.strong")[row]
    list.append({'name': col.text})

with open('DB/b3.json', 'w') as f:
    json.dump(list, f)
```

```python
with open("DB/b3.json", "r") as f:
    data = json.load(f)

list_yf = []
list_error = []

for i in range(0, len(data)):
    tries = 0
    while (tries < 5):
        try:
            ticker = yf.download(data[i]["name"] + ".SA", period="max", interval="1d")
            if (ticker.empty):
                list_error.append(data[i]["name"])
                break
            market_cap = ticker["Volume"].mean() * ticker["Close"].mean()
            list_yf.append({data[i]["name"] : [{'date': ticker.reset_index()['Date'][0].date().isoformat()}, {'volume_price_mean': market_cap}]})
        except RuntimeError:
            continue
        except KeyError:
            list_error.append(data[i]["name"])
            continue
        break


with open('DB/{}-b3.json'.format(date.today()), 'w') as f:
    json.dump(list_yf, f)

print(list_error)
```
